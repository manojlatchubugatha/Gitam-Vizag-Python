### np.zeros() np.ones()
- To initialize the weights during the first iteration of TensorFlow

```python
x = np.zeros((2,2)) # dtype as float
print(x)
```

```python
x = np.zeros((2,2),dtype=np.int16)
print(x)
```

```python
x2 = np.ones((3,3))
print(x2)
```

```python
x2 = np.arange(15)
print(x2)
```

```python
# Convert the single dimensional to two dimensional array
x3 = x2.reshape(5,3)
print(x3)
```

```python
x4 = x3 + 10
print(x4)
```

### Visualizing Data

- An interesting way to display Dataframe
- With matplotlib

```python
import matplotlib.pyplot as plt
a1 = [1,2,3,4]
a2 = [5,6,7,8]
plt.figure(figsize=(5,5))
plt.plot(a1,a2,color='red',marker='*')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
import numpy as np
divisions = ['A','B','C','D']
marks = [79,82,75,63]
avg_marks = [56,72,78,89]
index = np.arange(len(divisions))
width = 0.70
plt.bar(divisions,marks,color="red")
```

```python
lang = ['Python', 'Java', 'C','C++']
rating = [75,70,55,65]
plt.pie(rating,labels = lang,shadow=True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.show()
```
